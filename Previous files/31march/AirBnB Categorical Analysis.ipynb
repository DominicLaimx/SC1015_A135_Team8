```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
def concatcsv(inputfile,output):
    tempdata = pd.read_csv(inputfile)
    tempdata['city'] = inputfile.split('_')[0]
    tempdata['weekdays'] = inputfile.split('_')[1].split('.')[0] == 'weekdays'
    output = pd.concat([output,tempdata],axis=0)
    return output
```

```python
data = pd.read_csv('amsterdam_weekdays.csv',header='infer')
data['city']='amsterdam'
data['weekdays']=1
data.head()
```

```python
def outlier(df, col):
    sorted(df[col])
    Q1,Q3 = np.percentile(df[col] , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
lower , upper = outlier(data, 'realSum')
print("lower: ",lower)
print("upper: ",upper)
```

```python
outlierindex = data[data['realSum']>upper].index
data.drop(index=outlierindex, inplace=True)
```

# Relationship of variables(Categorical)

```python
#function to remove outliers
def remove_outliers(column,frame):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1
    new_df = frame[(frame[column] <= Q3+1.5*IQR) & (frame[column] >= Q1-1.5*IQR)]
    return new_df
```
